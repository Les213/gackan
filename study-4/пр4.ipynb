```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

```

```python
df = pd.read_csv('vgsales_14.csv')
df
```

```python
df.where(df['Rank'] == 10856).dropna()
```

```python
df = df.replace({'Name' : 'FIFA Street 3'}, 'Bleach')
df
```

```python
simp = SimpleImputer(missing_values=np.nan)
simp = simp.fit(df[['Rank']])

df['Rank'] = simp.transform(df[['Rank']])
df.isnull().mean()*100
```

```python
df.to_csv('newDF.csv', index=False)
```
